# Assignment 1 (Natural Language Processing)

### Importing Packages 

```python
from nltk import word_tokenize, sent_tokenize
from operator import itemgetter
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import random
```

### Calculating $TTR$
We assume that punctautions are not tokens.

```python
#Opening File
text_file = open('74-0.txt','r').read().decode('utf-8')
words_current = word_tokenize(text_file)
words = []
vocab = {}
punctuations = set('?/.>,<"|\;:]}[{~`!@''5#$%^&*()_-').union(set("'"))

#calculating Tokens and Types
for word in words_current:
    if len(set(word).intersection(punctuations)) == 0 and (len(word)>1 or word == 'a' or word == 'A' or word == 'I' or word == 'O') :
        words.append(word)
        if word in vocab:
            vocab[word] += 1
        else:
            vocab[word] = 1

tokens = len(words)
types = len(vocab)
ttr = float(types)/tokens
print 'Number of tokens = ' + str(tokens)
print 'Number of types = ' + str(types)
print 'TTR = ' + str(ttr) + '\n'
```

### Verifying Zipf's Law ($f.r = k$)

```python
print 'Printing Word, Frequency, Rank, f.r, Length_of_word, for top 100 word types...'
rank = 1
freq = []
ranks = []
length = []
wordsl = []
sorted_vocab = reversed(sorted(vocab.items(), key = itemgetter(1)))
count = 0
for key, value in sorted_vocab:
    freq.append(value)
    ranks.append(rank)
    length.append(len(key))
    wordsl.append(key)
    rank+=1
    count +=1
    if count <=100:
        print key,'\t', value, '\t', rank, '\t', value*rank, '\t', len(key)

#Function of Zipf's Law and Zipf's Law of length
def func(x, k):
    return np.divide(k,x)

#Calculating Parameter for Zipf's Law
params, params_covariance = optimize.curve_fit(func, ranks, freq,p0=[2])
print "\nParameter of Zipf's Law, k = " + str(params[0])
A = params[0]/types
print "A for this text = " + str(A)

#Plotting the data and function of Zipf's Law
fig = plt.figure(figsize = (12, 8))
plt.plot(ranks, func(ranks, params[0]),label="Zipf's Law", color='red')
plt.scatter(ranks, freq, label = 'Data')
plt.legend(loc='best')
plt.ylabel("Frequency")
plt.xlabel("Rank")
fig.suptitle("Zipf's Law")
plt.show()
```

### Verifying Zipf's Law of length ($f.l = c$)

```python
#Calculating Parameter(c) for Zipf's Law of length
params2, params_covariance2 = optimize.curve_fit(func, freq, length,p0=[1])
print "\nParameter of Zipf's Law of length, c = " + str(params2[0])

#Plotting the data and function of Zipf's Law
fig = plt.figure(figsize = (10,8))
plt.plot(freq, func(freq, params2[0]),label="Zipf's Law of length", color='red')
plt.scatter(freq, length, label = 'Data')
plt.legend(loc='best')
plt.xlabel("Frequency")
plt.ylabel("Length of word")
fig.suptitle("Zipf's Law of length")
plt.show()
```

We notice that Zipf's Law of length does not work for words with less frequency and short lengths. It works for words with longer length with less frequency and shorter length with high frequency. 

### Verifying Zipf's Law of meaning ($m = k \sqrt f$)

```python
#Choosing random words
np.random.seed(10000)
num_arr = [np.random.randint(1, types) for i in range(20)]
num_arr.sort()
freqm = []
meanings = []
wordsm =[]
for i in range(20):
    freqm.append(freq[num_arr[i]])
    wordsm.append(wordsl[num_arr[i]])
print 'List of randomly selected words is:\n', wordsm
print 'The corresponding frequencies of the words are:\n', freqm
print 'The corresponding number of meanings of the words in the text are:'
meanings = [9,3,2,2,3,2,1,2,2,2,1,2,2,1,1,1,1,1,1,1]
print meanings

#Function of Zipf's Law of meaning
def funcm(x, k):
    return k * np.sqrt(x)

#Calculating Parameter(k) for Zipf's Law of meaning
params4, params_covariance4 = optimize.curve_fit(funcm, freqm, meanings,p0=[1])
print "\nParameter of Zipf's Law of meaning, k = " + str(params4[0])

#Plotting the data and function of Zipf's Law
fig = plt.figure(figsize = (6,8))
plt.plot(freqm, funcm(freqm, params4[0]),label="Zipf's Law of meaning", color='red')
plt.scatter(freqm, meanings, label = 'Data')
plt.legend(loc='best')
plt.xlabel("Frequency")
plt.ylabel("Number of meanings")
fig.suptitle("Zipf's Law of meaning")
plt.show()
```

### Verifying Heaps' Law ($|V| = KN^\beta$)

```python
documents = sent_tokenize(text_file)

vocabulary = []
word_count = 0 
x = []
y = []
for sent in documents:
    current_words = word_tokenize(sent)
    for word in current_words:
        if len(set(word).intersection(punctuations)) == 0 and (len(word)>1 or word == 'a' or word == 'A' or word == 'I' or word == 'O') :
            word_count+=1
            if word not in vocabulary:
                vocabulary.append(word)
    x.append(word_count)
    y.append(len(vocabulary))
    
#Function of Heaps' Law
def funch(x, k, b):
    return k * np.power(x, b)

#Calculating Parameters for Heaps' Law 
params3, params_covariance3 = optimize.curve_fit(funch, x, y,p0=[1, 1])
print "\n Parameters of Heaps' Law are:\n k = " + str(params3[0]) + "\n beta = " + str(params3[1])

#Plotting the data and Heaps' Law
fig = plt.figure(figsize = (10,8))
plt.plot(x, funch(x, params3[0], params3[1]),label="Heaps' Law", color='red')
plt.scatter(x, y, label = 'Data', s = 2)
plt.legend(loc='best')
plt.xlabel("Number of tokens")
plt.ylabel("Size of vocabulary")
fig.suptitle("Heaps' Law")
plt.show()
```
